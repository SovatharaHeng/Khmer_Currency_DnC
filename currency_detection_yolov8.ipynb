```
from ultralytics import YOLO
from IPython.display import Image, display
import os
import shutil

# Load model
model = YOLO('/content/yolov8n.pt')

# Make sure folders exist
os.makedirs('test_images', exist_ok=True)
os.makedirs('/content/predictions', exist_ok=True)

# Image to predict
image_path = 'test_images/test1.jpg'

# Run prediction (save=True writes to runs/detect/predict)
results = model.predict(source=image_path, save=True)

# Get filename only
image_name = os.path.basename(image_path)

# Define source and destination paths
source_path = f'runs/detect/predict/{image_name}'
destination_path = f'/content/predictions/{image_name}'

# Move the result image
shutil.move(source_path, destination_path)

# Display the saved result
display(Image(filename=destination_path))
```
